```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

#Generate Data Using a Dummy function
X = np.arange(10)
Y = (X-5) ** 2 + 3

print(X,Y)
```

```python
#Visualise
plt.plot(X, Y)
plt.show()
```

```python
# Function minimum value is 3 when x = 5 

#Gradient Descent

#init x with any random value
x = 9
lr = 0.1

# formula x = x - lr * dy / dx
# f(x) = (x -5) ^ 2 + 3 
# dy / dx = 2(x-5)

for i in range(50):
    grad =  2 * ( x - 5)
    x = x - lr * grad
    y = (x-5) ** 2 + 3
    print(x,y)
    plt.scatter(x,y)
    
plt.plot(X, Y)
plt.show()
```
